##**Introducción**
En el presente trabajo se realiza un análisis estadístico de los datos correspondientes a los nacimientos en Colombia durante el año 2023, tomando como base la información suministrada por el Departamento Administrativo Nacional de Estadística (DANE). El propósito central es analizar el comportamiento del tiempo de gestación y de la talla al nacer en los nacidos vivos del país, describiendo cómo se distribuye cada variable a partir de gráficos, proporciones y medidas estadísticas. Asimismo, se busca evaluar si existe algún tipo de dependencia o asociación entre ambas variables, apoyándose en la tabla de contingencia, las probabilidades condicionales y la prueba de Chi-cuadrado. De esta manera, se pretende ofrecer una visión clara y fundamentada sobre el comportamiento de las dos variables y su posible relación dentro de la población analizada.






##**Acerca de las Variables**

**VARIABLE: TALLA_NAC**
Esta variable recoge la talla que presenta un bebé al momento de nacer. Se trata de una variable cualitativa discreta. Su importancia radica en que la medida de la talla es uno de los indicadores más utilizados para valorar cómo se dio el crecimiento del bebé dentro del útero y para identificar posibles señales de alerta en su estado de salud. Además, esta información es útil para análisis de salud pública, ya que permite estudiar factores maternos, sociales y ambientales que pueden influir en el desarrollo infantil.

**VARIABLE: T_GES**
Representa el tiempo de gestación con el que ocurre el nacimiento. También corresponde a una variable cualitativa discreta. Este dato es fundamental porque permite distinguir si el nacimiento fue prematuro, a término o después de lo esperado, lo cual tiene implicaciones directas en la salud del recién nacido. El tiempo de gestación se utiliza además para evaluar el proceso de cuidado prenatal y para orientar acciones médicas cuando se identifican riesgos relacionados con la duración del embarazo.

# Tabla de contingencia


*   Representación gráfica de una tabla de contingencia
*   Proporciones a una tabla de contingencia



```python
df2.loc[df2.TALLA_NAC == '1' , "TALLA_NAC"] = "<20"
df2.loc[df2.TALLA_NAC == '2', "TALLA_NAC"] = "20-29"
df2.loc[df2.TALLA_NAC == '3', "TALLA_NAC"] = "30-39"
df2.loc[df2.TALLA_NAC == '4', "TALLA_NAC"] = "40-49"
df2.loc[df2.TALLA_NAC == '5', "TALLA_NAC"] = "50-59"
df2.loc[df2.TALLA_NAC == '6', "TALLA_NAC"] = ">60" # Corrected typo, previously targeted T_GES
df2.loc[df2.T_GES == '1', "T_GES"] = "<21"
df2.loc[df2.T_GES == '2', "T_GES"] = "22-27"
df2.loc[df2.T_GES == '3', "T_GES"] = "28-37"
df2.loc[df2.T_GES == '4', "T_GES"] = "38-41"
df2.loc[df2.T_GES == '5', "T_GES"] = ">42"
df2.loc[df2.T_GES == '6', "T_GES"] = "IGNORADOS"
df2.T_GES, df2.TALLA_NAC = df2.T_GES.astype(str), df2.TALLA_NAC.astype(str)
df2.T_GES = df2.T_GES.astype(str)
df2.head(3)

```

```python
tabla_cont = pd.crosstab(df2.T_GES, df2.TALLA_NAC)
tabla_cont
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,6))
heatmap = sns.heatmap(tabla_cont, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.show()
```

Con la tabla de contingencia se puede realizar las probabilidades, que se pueden obtener. En este caso, n= 508827, vamos a calcular el tiempo de gestación de 38 a41 semanas según el tamaño del recién nacido que es 40 a 49; la probabilidad es p:140106/508827 = 27,53%.


A continuación se realizará la prueba de Chi-cuadrado, consiste en que la hipótesis nula corresponde a que las variables:
Ho: X y Y no están asociadas, o son independientes o son homogéneas

*   Ho: X y Y no están asociadas, X y Y son independientes, X y Y son homogéneas
*   Ha: X y Y estén asociadas, Y dependa de X, X y Y no son homogéneas



```python
res=stats.chi2_contingency(tabla_cont)
chi, valor_P, _, _ = res
print(f"El valor de la estadistica Chi-cuadrado es {chi} y el valo", chi)
print("Valor P:", valor_P)
```



*   si valor p < 0.05 se rechaza la hipotesis nula
*   si valor p > 0.05 no se rechaza la hipotesis nula
como el valor p $=0.0 < 0.05se rechaza la hipotesis nula, es decir, se rechaza con una confianza del 95% que talla del recien nacido y tiempo de gestacion del recien nacido estan asociadas
